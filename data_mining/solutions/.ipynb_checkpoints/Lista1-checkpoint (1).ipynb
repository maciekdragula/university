```python
# **************** Data ****************
mu = 5
sigma = 3
data1 = np.random.normal(mu, sigma, 10000)
data2 = np.random.normal(mu, sigma, 100000)
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
plt.figure(figsize=(20, 5))

# **************** Histogram 10k samples ****************
plt.subplot(1, 3, 1)
plt.hist(data1, 100)
plt.xlabel('X~N[-1,1]')
plt.ylabel('N')
plt.title("Normal distribution - histogram (10k samples)")
plt.axis([mu - 4 * sigma, mu + 4 * sigma, 0, 400])
plt.grid(True)

#PDF
plt.subplot(1, 3, 2)
plt.xlabel('X')
plt.ylabel('N[-1,1]')
plt.title("Normal distribution - PDF")
plt.axis([mu - 4 * sigma, mu + 4 * sigma, 0, 0.16])
plt.grid(True)
plt.plot(x, sstats.norm.pdf(x, mu, sigma), color='red')

# **************** Histogram 100k samples ****************
plt.subplot(1, 3, 3)
plt.hist(data2, 100, color='green')
plt.xlabel('X~N[-1,1]')
plt.ylabel('N')
plt.title("Normal distribution - histogram (100k samples)")
plt.axis([mu - 4 * sigma, mu + 4 * sigma, 0, 4000])
plt.grid(True)


plt.show()
```

Skorzystamy z nastpjcych faktw: <br>
- $E[aX + bY] = aE[X] + bE[X]$ <br>
- $Var(a \cdot X) = a^2 \cdot Var(X)$ <br>
- $Var(X + Y) = Var(X) + Var(Y)$ (gdy zmienne losowe $X$ i $Y$ s liniowo niezalene) <br>
- Kombinacja liniowa rozkadw normalnych jest rozkadem normalnym. <br>
<a/>
<br>

Zdefiniujmy zmienn losow $Z = Y - X$, gdzie $Y \sim \mathcal{N}(3, 1)$ i $X \sim \mathcal{N}(2, 5)$. Wwczas $Z$ jest zmienn losow z rozkadem normalnym z wartoci oczekiwan $E[Z] = E[Y - X] = E[Y] - E[X] = 3 - 2 = 1$ i wariancj $Var(Z) = Var(Y - X) = Var(Y) + (-1)^2 Var(X) = 1 + 25 = 26$. <br>
<br>
Chcielibymy wiedzie, ile wynosi $P(Z > 0) = 1 - P(Z \le 0)$. Dystrybuanta zmiennej losowej z rozkadem normalnym $P(X \le x)$ jest wyraona przez nieelementar funkcj $\Phi(\frac{x - \mu}{\sigma})$, ktrej warto mona odczyta z tablic. Wwczas:
$$ P(X > Y) = 1 - P(Z \le 0) = 0.5777403662635300 $$

```python
def v_len(x):
    return np.sqrt(np.dot(x.T, x)).item()

def weighted_avg(x, w):
    n = len(x)
    return (np.dot(x.T, w)/n).item()
        
def euclid_dist(x, y):
    d = x - y
    return np.sqrt(np.dot(d.T, d)).item()
        
def dot_product(x, y):
    return np.dot(x.T, y).item()

x = np.random.rand(100, 1)
w = np.random.rand(100, 1)
y = np.random.rand(100, 1)

print(v_len(x))
print(weighted_avg(x, w))
print(euclid_dist(x, y))
print(dot_product(x, y))
```

```python
def dist_point_to_line(p, A, B, C):
    if len(p) != 2:
        raise "x must have exactly two cooridinates!"
    x, y = p
    return abs(A * x + B * y + C)/np.sqrt(A**2 + B**2)
    
```

```python
d = np.apply_along_axis(lambda x: dist_point_to_line(x, 2, -1, 5), 1, data)
print(d)
print(sum(d))
```

```python
d_better = np.apply_along_axis(lambda x: dist_point_to_line(x, 2, -1, 6), 1, data)
print(d_better)
print(sum(d_better))
```
