Skorzystamy z następjących faktów: <br>
- $E[aX + bY] = aE[X] + bE[X]$ <br>
- $Var(a \cdot X) = a^2 \cdot Var(X)$ <br>
- $Var(X + Y) = Var(X) + Var(Y)$ (gdy zmienne losowe $X$ i $Y$ są liniowo niezależne) <br>
- Kombinacja liniowa rozkładów normalnych jest rozkładem normalnym. <br>
<a/>
<br>

Zdefiniujmy zmienną losową $Z = Y - X$, gdzie $Y \sim \mathcal{N}(3, 1)$ i $X \sim \mathcal{N}(2, 5)$. Wówczas $Z$ jest zmienną losową z rozkładem normalnym z wartością oczekiwaną $E[Z] = E[Y - X] = E[Y] - E[X] = 3 - 2 = 1$ i wariancją $Var(Z) = Var(Y - X) = Var(Y) + (-1)^2 Var(X) = 1 + 25 = 26$. <br>
<br>
Chcielibyśmy wiedzieć, ile wynosi $P(Z > 0) = 1 - P(Z \le 0)$. Dystrybuanta zmiennej losowej z rozkładem normalnym $P(X \le x)$ jest wyrażona przez nieelementarą funkcję $\Phi(\frac{x - \mu}{\sigma})$, której wartość można odczytać z tablic. Wówczas:
$$ P(X > Y) = 1 - P(Z \le 0) = 0.5777403662635300 $$

```python
def v_len(x):
    return np.sqrt(np.dot(x.T, x)).item()

def weighted_avg(x, w):
    n = len(x)
    return (np.dot(x.T, w)/n).item()
        
def euclid_dist(x, y):
    d = x - y
    return np.sqrt(np.dot(d.T, d)).item()
        
def dot_product(x, y):
    return np.dot(x.T, y).item()

x = np.random.rand(100, 1)
w = np.random.rand(100, 1)
y = np.random.rand(100, 1)

print(v_len(x))
print(weighted_avg(x, w))
print(euclid_dist(x, y))
print(dot_product(x, y))
```

```python
def dist_point_to_line(p, A, B, C):
    if len(p) != 2:
        raise "x must have exactly two cooridinates!"
    x, y = p
    return abs(A * x + B * y + C)/np.sqrt(A**2 + B**2)
    
```

```python
d = np.apply_along_axis(lambda x: dist_point_to_line(x, 2, -1, 5), 1, data)
print(d)
print(sum(d))
```

```python
d_better = np.apply_along_axis(lambda x: dist_point_to_line(x, 2, -1, 6), 1, data)
print(d_better)
print(sum(d_better))
```
