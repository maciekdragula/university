```python
# **************** Data ****************
data1 = np.random.uniform(-1, 1, 10000)
data2 = np.random.uniform(-1, 1, 100000)
x = np.linspace(-1, 1, 100)

fig, ax1 = plt.subplots(figsize=(8, 5))

# **************** Histogram 10k samples ****************
ax1.set_xlabel('$X \sim \mathcal{U}[-1,1]$', fontsize=13)
ax1.set_ylabel('$N$ (100k samples)', fontsize=13)
ax1.set_xlim(-1.1, 1.1)
ax1.set_ylim(0, 1400)
ax1.hist(data2, 100, alpha=1, color='teal', edgecolor='black')
ax1.tick_params(axis='y', labelcolor='teal')

# **************** Histogram 100k samples ****************
ax2 = ax1.twinx()

ax2.set_ylabel('$N$ (10k samples)', fontsize=13)
ax2.set_ylim(0, 140)
ax2.hist(data1, 100, alpha=0.85, color='chocolate', edgecolor='black')
ax2.tick_params(axis='y', labelcolor='chocolate')

#PDF
ax3 = ax1.twinx()

ax3.set_ylim(0, 0.7)
ax3.plot(x, sstats.uniform.pdf(x, -1, 2), color='red')
ax3.get_yaxis().set_visible(False)

fig.tight_layout()
plt.title("Uniform distribution - histogram and PDF", fontsize=13)
plt.show()
```

```python
# **************** Data ****************
mu = 5
sigma = np.sqrt(3)
data1 = np.random.normal(mu, sigma, 10000)
data2 = np.random.normal(mu, sigma, 100000)
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
fig, ax1 = plt.subplots(figsize=(8, 5))

# **************** Histogram 100k samples ****************
ax1.set_xlabel('$X \sim \mathcal{N}[-1,1]$', fontsize=13)
ax1.set_ylabel('$N$ (100k samples)', fontsize=13)
ax1.set_xlim(mu - 4 * sigma, mu + 4 * sigma)
ax1.set_ylim(0, 4000)
ax1.hist(data2, 100, alpha=1, color='teal', edgecolor='black')
ax1.tick_params(axis='y', labelcolor='teal')

# **************** Histogram 10k samples ****************
ax2 = ax1.twinx()

ax2.set_ylabel('$N$ (10k samples)', fontsize=13)
ax2.set_ylim(0, 400)
ax2.hist(data1, 100, alpha=0.85, color='chocolate', edgecolor='black')
ax2.tick_params(axis='y', labelcolor='chocolate')

#PDF
ax3 = ax1.twinx()

ax3.set_ylim(0, 0.257)
ax3.plot(x, sstats.norm.pdf(x, mu, sigma), color='red')
ax3.get_yaxis().set_visible(False)

fig.tight_layout()
plt.title("Normal distribution - histogram and PDF", fontsize=13)
plt.show()
```

Skorzystamy z nastpjcych faktw: <br>
- $E[aX + bY] = aE[X] + bE[X]$ <br>
- $Var(a \cdot X) = a^2 \cdot Var(X)$ <br>
- $Var(X + Y) = Var(X) + Var(Y)$ (gdy zmienne losowe $X$ i $Y$ s liniowo niezalene) <br>
- Kombinacja liniowa rozkadw normalnych jest rozkadem normalnym. <br>
<a/>
<br>

Zdefiniujmy zmienn losow $Z = Y - X$, gdzie $Y \sim \mathcal{N}(3, 1)$ i $X \sim \mathcal{N}(2, 5)$. Wwczas $Z$ jest zmienn losow z rozkadem normalnym z wartoci oczekiwan $E[Z] = E[Y - X] = E[Y] - E[X] = 3 - 2 = 1$ i wariancj $Var(Z) = Var(Y - X) = Var(Y) + (-1)^2 Var(X) = 1 + 5 = 6$. <br>
<br>
Chcielibymy wiedzie, ile wynosi $P(Z > 0) = 1 - P(Z \le 0)$. Dystrybuanta zmiennej losowej z rozkadem normalnym $P(X \le x)$ jest wyraona przez nieelementar funkcj $\Phi(\frac{x - \mu}{\sigma})$, ktrej warto mona odczyta z tablic. Wwczas:
$$ P(X > Y) = 1 - P(Z \le 0) = 0.6584543008451956 $$

```python
def v_len(x):
    return np.sqrt(np.dot(x.T, x)).item()

def weighted_avg(x, w):
    n = len(x)
    return (np.dot(x.T, w)/n).item()
        
def euclid_dist(x, y):
    d = x - y
    return np.sqrt(np.dot(d.T, d)).item()
        
def dot_product(x, y):
    return np.dot(x.T, y).item()

x = np.random.rand(100, 1)
w = np.random.rand(100, 1)
y = np.random.rand(100, 1)

print(v_len(x))
print(weighted_avg(x, w))
print(euclid_dist(x, y))
print(dot_product(x, y))
```

# Zadanie 4

```python
def f(x):
    return 2 * x - 8
```

```python
def create_df(df, f):
    labels = []
    
    for index, row in df.iterrows():
        if row['species'] == 'setosa':
            if row['sepal width (cm)'] >= f(row['sepal length (cm)']):
                labels.append(1)
            else:
                labels.append(0)
        elif row['species'] == 'versicolor':
            if row['sepal width (cm)'] <= f(row['sepal length (cm)']):
                labels.append(1)
            else:
                labels.append(0)
                
    new_df = df.query('species == "setosa" or species == "versicolor"')
    new_df = new_df.assign(labels=labels)
    return new_df
```

```python
bounded_df1 = create_df(iris_df, f)

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='labels', palette=['r', 'g'], data=bounded_df1, legend=False)
plt.title('Iris')
plt.tick_params(axis='both', direction="in")
plt.xlim(3, 9)
plt.xticks(np.arange(3, 10, 1))
plt.ylim(1, 5, 1)
plt.yticks(np.arange(1, 6, 1))
x = np.linspace(4, 7, 20)
plt.plot(x, f(x), color='navy')
plt.savefig('zadanie2a.png')
plt.show()
```

```python
def g(x):
    return 1.3 * x - 3.8
```

```python
bounded_df2 = create_df(iris_df, g)
assert all(bounded_df2['labels'] == 1)

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='labels', palette=['g'], data=bounded_df2, legend=False)
plt.title('Iris')
plt.tick_params(axis='both',direction="in")
plt.xlim(3, 9)
plt.xticks(np.arange(3, 10, 1))
plt.ylim(1, 5, 1)
plt.yticks(np.arange(1, 6, 1))
x = np.linspace(3, 7, 20)
plt.plot(x, g(x), color='navy')
plt.savefig('zadanie2b.png')
plt.show()
```

# Zadanie 6

```python
def calc_coordinates(k, d):
    R = d / (2 * np.sin(np.pi / k))
    angle = 2 * np.pi / k
    XX = []
    YY = []
    
    for i in range(k):
        x = R * np.sin(i * angle)
        y = R * np.cos(i * angle)
        XX.append(x)
        YY.append(y)
    
    return XX, YY

def gen_clouds(XX, YY):
    df = pd.DataFrame(columns=['x', 'y', 'cloud'])
    data_x = []
    data_y = []
    cloud = []
    i = 0
    
    for x, y in zip(XX, YY):
        data_x = np.concatenate((data_x, np.random.normal(x, 1, 1000)))
        data_y = np.concatenate((data_y, np.random.normal(y, 1, 1000)))
        cloud += 1000*[i]
        i += 1
    
    df['x'] = data_x
    df['y'] = data_y
    df['cloud'] = cloud
    return df
```

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
K = [7, 11, 23]
D = [5, 10, 15]

for i in range(3):
    k = K[i]
    for j in range(3):
        d = D[j]
        
        XX, YY = calc_coordinates(k, d)
        df = gen_clouds(XX, YY)
        sns.scatterplot(x='x', y='y', hue='cloud', data=df, legend=False, ax=axes[i, j])

plt.show()    
```

# Zadanie 7

```python
def calc_cloest_center(df, XX, YY):
    cloud = df.loc[:, ['x', 'y']].to_numpy()
    centres = np.array([XX, YY])

    dists = np.sum(cloud**2, axis=1).reshape(-1, 1) + np.sum(centres**2, axis=0).reshape(1, -1)
    dists -= 2 * np.dot(cloud, centres)
    
    cloest = np.argmin(dists, axis=1)
    return cloest
    

def extend_df(df, XX, YY):
    cloest = calc_cloest_center(df, XX, YY)
    check = (df.loc[:, 'cloud'] == cloest).astype(int)
    df = df.assign(cloest_correct=check)
    return df
```

```python
wrong_on_top = True

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
K = [7, 11, 23]
D = [5, 10, 15]

for i in range(3):
    k = K[i]
    for j in range(3):
        d = D[j]

        XX, YY = calc_coordinates(k, d)
        df = gen_clouds(XX, YY)
        df_extended = extend_df(df, XX, YY)
        
        if wrong_on_top:
            df_extended = df_extended.sort_values(by=['cloest_correct'], ascending=False)
        
        palette = ['g'] if all(df_extended['cloest_correct'] == 1) else ['r', 'g']
        sns.scatterplot(x='x', y='y', hue='cloest_correct', data=df_extended, palette=palette, legend=False, ax=axes[i, j])
        
plt.show()
```

```python
for i in range(3):
    print("Group #%d..." % i)
    res = []
    for k in [4, 5, 7, 9, 11, 14, 16, 20, 23]:
        errs = []
        test_d = np.arange(0.1, 10, 0.1)
        for d in test_d:
            XX, YY = calc_coordinates(k, d)
            df = gen_clouds(XX, YY)
            df_extended = extend_df(df, XX, YY)
            err = 1 - np.sum(df_extended.loc[:, 'cloest_correct']) / (k * 1000)
            errs.append(abs(0.1 - err))
        best_d = np.argmin(errs)
        res.append(round(test_d[best_d], 3))
    print(res)
```
