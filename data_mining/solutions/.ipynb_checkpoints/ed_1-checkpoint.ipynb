```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

#Parameters to set
mu_x = 2
variance_x = 5

mu_y = 3
variance_y = 1

# X, Y = math.sqrt(5) * r.randn(N) + 2, math.sqrt(1) * r.randn(N) + 3

#Create grid and multivariate normal
x = np.linspace(min(xpos),max(xpos),500)
y = np.linspace(min(ypos),max(ypos),500)

X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.bar3d(xpos, ypos, zpos, dx, dy, dz)
plt.show()
fig2 = plt.figure()
ax2 = fig2.gca(projection='3d')
ax2.plot_surface(X, Y, rv.pdf(pos),linewidth=0, alpha=0.8,color='red')
plt.show()
```

```python
# zadanie 4
from sklearn import datasets 
import matplotlib.pyplot as plt

iris = datasets.load_iris() 



colors = ['r', 'g']
def f(a, b):
    plt.title("Iris")
    plt.ylabel('sepal width (cm)')
    plt.xlabel('sepal length (cm)')

    for i in range(2):

        data = iris.data[iris.target == i]
        above_line = data[:,1] > data[:,0]*a + b
        below_line = [not el for el in above_line]

        green, red = above_line, below_line
        if i == 1:
            green, red = red, green

        plt.scatter(data[:,0][green], data[:,1][green], c = 'green')
        plt.scatter(data[:,0][red], data[:,1][red], c = 'red')         

    plt.axis([3, 9, 1, 5])
    plt.yticks([i for i in range(1,6)])

    # b)
    plt.plot([3, 9], [a*3+b, a*9+b], '-')

    plt.savefig("zadanie2a.png")
    plt.show()

f(2,-8)

# e) 
def find_function():
    for a in range(10, 20, 1):
        for b in range(-100, 100, 1):
            wrong = False
            for i, e in enumerate(iris.data):
                if ((e[1] < e[0]*a/10 + b/10 and iris.target[i] == 0)
                    or (e[1] > e[0]*a/10 + b/10 and iris.target[i] == 1)):
                    wrong = True
                    break
            if wrong is False:
                return a/10,b/10

a, b = find_function()
print(a,b)
f(a,b)
```

```python
# zadanie 5
import math

points = [(1, 10), (2, 10),
(3, 11), (4, 12), (5, 18), (6, 18), (7, 19), (8, 26), (9, 19), (10, 26)]

transformed = ([], [])
for x, y in points:
    transformed[0].append(x)
    transformed[1].append(y)
plt.scatter(transformed[0], transformed[1])

def plot_line(a,b):
    plt.plot([0, 12], [a*0+b, a*12+b], '-')


plot_line(2,5)
plt.savefig("zadanie3a.png")

def eukl(a, b, x,y):
    return abs(a*x + -y + b)/math.sqrt(a**2 + 1)

def find_function():
    dists = [eukl(2,5,x,y) for (x,y) in points]
    # print(dists)

    best_sum, best_par = sum(dists), (2,5)
    print("first sum", best_sum)
    for a in range(-1000, 1000, 1):
        for b in range(-1000, 1000, 1):
            wrong = False
            s = sum([eukl(a/10,b/10,x,y) for (x,y) in points])
            if s < best_sum:
                best_sum = s
                best_par = (a/10,b/10)
    print("better sum", best_sum)
    return best_par

a, b = find_function()
print(a,b)
plot_line(a,b)
plt.show()
```

```python
# zadanie 7
import numpy as np
import math
import matplotlib.pyplot as plt
N = 1000
for d in [5, 10, 15]:
    for K in [7, 11, 23]:
        d = K * 0.53
        print("K =", K, "d =", d)
        a, b = np.sin(2*math.pi/K * np.arange(K)) * d, np.cos(2*math.pi/K * np.arange(K)) * d        
        
        points, good = [], []
        for i in range(K):
            cloud = np.random.multivariate_normal([a[i], b[i]], [[1,0], [0,1]], N).T
            
            points.append(cloud)
            distsI = (cloud[0] - a[i])**2 + (cloud[1] - b[i])**2
            
            left = i - 1 if i > 0 else K - 1
            right = i + 1 if i < K-1 else 0
            distsL, distsR = [(cloud[0] - a[j])**2 + (cloud[1] - b[j])**2 for j in [left, right]]
            distsLR = np.minimum(distsL, distsR)
            good.append(distsI <= distsLR)
            
        print(round(100-np.sum(good)/(N*K)*100,2), "%")
        for i, p in enumerate(points):
            
            plt.scatter(p[0][good[i]],p[1][good[i]],color='green')
            plt.scatter(p[0][good[i] == False],p[1][good[i] == False],color='red')

        plt.show()
    # break
```
