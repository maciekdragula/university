# Zadanie 1

### a)

```python
mu = np.array([3, 5])
sigma = np.array([[12, 3], [3, 1]])

X = np.random.multivariate_normal(mu, sigma, 1000)
```

### b)

```python
X_std = (X - X.mean(axis=0)) / X.std(axis=0)
```

### c)

```python
S = np.cov(X_std.T)

print(S)
```

### d)

```python
lambdas, vectors = np.linalg.eig(S)
print(lambdas)
print(vectors)
```

### e)

```python
X_pca = np.dot(X_std, np.dot(vectors, np.diag(1 / np.sqrt(lambdas))))
```

### f)

```python
x_left_top_raw = X.mean(axis=0)[0] - mu[0] * vectors[0][0] * X.std(axis=0)[0]
x_right_bottom_raw = X.mean(axis=0)[0] + mu[0] * vectors[0][0] * X.std(axis=0)[0]

y_left_top_raw = X.mean(axis=0)[1] - mu[0] * vectors[0][1] * X.std(axis=0)[1]
y_right_bottom_raw = X.mean(axis=0)[1] + mu[0] * vectors[0][1] * X.std(axis=0)[1]


x_left_bottom_raw = X.mean(axis=0)[0] - mu[1] * vectors[1][0] * X.std(axis=0)[0]
x_right_top_raw = X.mean(axis=0)[0] + mu[1] * vectors[1][0] * X.std(axis=0)[0]

y_left_bottom_raw = X.mean(axis=0)[1] - mu[1] * vectors[1][1] * X.std(axis=0)[1]
y_right_top_raw = X.mean(axis=0)[1] + mu[1] * vectors[1][1] * X.std(axis=0)[1]
```

```python
x_left_top_std = X_std.mean(axis=0)[0] - mu[0] * vectors[0][0] * X_std.std(axis=0)[0]
x_right_bottom_std = X_std.mean(axis=0)[0] + mu[0] * vectors[0][0] * X_std.std(axis=0)[0]

y_left_top_std = X_std.mean(axis=0)[1] - mu[0] * vectors[0][1] * X_std.std(axis=0)[1]
y_right_bottom_std = X_std.mean(axis=0)[1] + mu[0] * vectors[0][1] * X_std.std(axis=0)[1]


x_left_bottom_std = X_std.mean(axis=0)[0] - mu[1] * vectors[1][0] * X_std.std(axis=0)[0]
x_right_top_std = X_std.mean(axis=0)[0] + mu[1] * vectors[1][0] * X_std.std(axis=0)[0]

y_left_bottom_std = X_std.mean(axis=0)[1] - mu[1] * vectors[1][1] * X_std.std(axis=0)[1]
y_right_top_std = X_std.mean(axis=0)[1] + mu[1] * vectors[1][1] * X_std.std(axis=0)[1]
```

```python
mu1 = np.array([-21, -2])
mu2 = np.array([3, 5])
mu3 = np.array([27, 12])
sigma = np.array([[12, 3], [3, 1]])

X_mix = [] 
for _ in range(1000):
    mu = random.randint(0, 2)
    if mu == 0:
        X_mix.append(np.random.multivariate_normal(mu1, sigma))
    elif mu == 1:
        X_mix.append(np.random.multivariate_normal(mu2, sigma))
    else:
        X_mix.append(np.random.multivariate_normal(mu3, sigma))
X_mix = np.array(X_mix)

X_mix_std = (X_mix - X_mix.mean(axis=0)) / X_mix.std(axis=0)

S_mix = np.cov(X_mix_std.T)

lambdas_mix, vectors_mix = np.linalg.eig(S_mix)

X_mix_pca = np.dot(X_mix_std, np.dot(vectors_mix, np.diag(1 / np.sqrt(lambdas_mix))))
```

```python
x_left_top_raw_mix = X_mix.mean(axis=0)[0] - 1 * vectors_mix[0][0] * X_mix.std(axis=0)[0]
x_right_bottom_raw_mix = X_mix.mean(axis=0)[0] + 1 * vectors_mix[0][0] * X_mix.std(axis=0)[0]

y_left_top_raw_mix = X_mix.mean(axis=0)[1] - 1 * vectors_mix[0][1] * X_mix.std(axis=0)[1]
y_right_bottom_raw_mix = X_mix.mean(axis=0)[1] + 1 * vectors_mix[0][1] * X_mix.std(axis=0)[1]


x_left_bottom_raw_mix = X_mix.mean(axis=0)[0] - 3 * vectors_mix[1][0] * X_mix.std(axis=0)[0]
x_right_top_raw_mix = X_mix.mean(axis=0)[0] + 3 * vectors_mix[1][0] * X_mix.std(axis=0)[0]

y_left_bottom_raw_mix = X_mix.mean(axis=0)[1] - 3 * vectors_mix[1][1] * X_mix.std(axis=0)[1]
y_right_top_raw_mix = X_mix.mean(axis=0)[1] + 3 * vectors_mix[1][1] * X_mix.std(axis=0)[1]
```

```python
x_left_top_std_mix = X_mix_std.mean(axis=0)[0] - 1 * vectors_mix[0][0] * X_mix_std.std(axis=0)[0]
x_right_bottom_std_mix = X_mix_std.mean(axis=0)[0] + 1 * vectors_mix[0][0] * X_mix_std.std(axis=0)[0]

y_left_top_std_mix = X_mix_std.mean(axis=0)[1] - 1 * vectors_mix[0][1] * X_mix_std.std(axis=0)[1]
y_right_bottom_std_mix = X_mix_std.mean(axis=0)[1] + 1 * vectors_mix[0][1] * X_mix_std.std(axis=0)[1]


x_left_bottom_std_mix = X_mix_std.mean(axis=0)[0] - 3 * vectors_mix[1][0] * X_mix_std.std(axis=0)[0]
x_right_top_std_mix = X_mix_std.mean(axis=0)[0] + 3 * vectors_mix[1][0] * X_mix_std.std(axis=0)[0]

y_left_bottom_std_mix = X_mix_std.mean(axis=0)[1] - 3 * vectors_mix[1][1] * X_mix_std.std(axis=0)[1]
y_right_top_std_mix = X_mix_std.mean(axis=0)[1] + 3 * vectors_mix[1][1] * X_mix_std.std(axis=0)[1]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(17, 4))
axes[0].scatter(X_mix[:, 0], X_mix[:, 1], edgecolors='black')
axes[0].plot([x_left_top_raw_mix, x_right_bottom_raw_mix], [y_left_top_raw_mix, y_right_bottom_raw_mix], '--', c='orange')
axes[0].plot([x_left_bottom_raw_mix, x_right_top_raw_mix], [y_left_bottom_raw_mix, y_right_top_raw_mix], '--', c='orange')
axes[0].set_title('Raw data')

axes[1].scatter(X_mix_std[:, 0], X_mix_std[:, 1], edgecolors='black')
axes[1].plot([x_left_top_std_mix, x_right_bottom_std_mix], [y_left_top_std_mix, y_right_bottom_std_mix], '--', c='orange')
axes[1].plot([x_left_bottom_std_mix, x_right_top_std_mix], [y_left_bottom_std_mix, y_right_top_std_mix], '--', c='orange')
axes[1].set_title('Standarized data')

axes[2].scatter(X_mix_pca[:, 0], X_mix_pca[:, 1], edgecolors='black')
axes[2].set_xlim([-4.5, 4.5])
axes[2].set_ylim([-3.5, 3.5])
axes[2].set_title('Data after PCA')

plt.show()
```

### j)

```python
iris_df = datasets.load_iris()

iris_X = np.array(iris_df.data)
iris_Y = np.array(iris_df.target).reshape(-1, 1)
```

```python
iris_X_std = (iris_X - iris_X.mean(axis=0)) / iris_X.std(axis=0)
iris_S_std = np.cov(iris_X_std.T)
iris_lambdas, iris_vectors = np.linalg.eig(iris_S_std)
iris_X_pca = np.dot(iris_X_std, np.dot(iris_vectors, np.diag(1 / np.sqrt(iris_lambdas))))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(17, 6))

axes[0].scatter(iris_X_std[:, 0], iris_X_std[:, 1], c=iris_Y.reshape(-1), edgecolors='black')
axes[0].set_title('Iris standarized')
axes[0].set_xlim([-2, 3])
axes[0].set_ylim([-2.5, 3.5])

axes[1].scatter(iris_X_pca[:, 0], iris_X_pca[:, 1], c=iris_Y.reshape(-1), edgecolors='black')
axes[1].set_title('Iris after PCA')
axes[1].set_xlim([-2, 3])
axes[1].set_ylim([-2.5, 3.5])

plt.show()
```

### k)

```python
iris_recreated = iris_X_pca * np.sqrt(iris_lambdas)
iris_recreated = np.dot(iris_recreated, iris_vectors)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(17, 6))

axes[0].scatter(iris_X_std[:, 0], iris_X_std[:, 1], c=iris_Y.reshape(-1), edgecolors='black')
axes[0].set_title('Original Iris (standarized)')
axes[0].set_xlim([-2, 3])
axes[0].set_ylim([-2.5, 3.5])

axes[1].scatter(iris_recreated[:, 0], iris_recreated[:, 1], c=iris_Y.reshape(-1), edgecolors='black')
axes[1].set_title('Recreated Iris')
axes[1].set_xlim([-2, 3])
axes[1].set_ylim([-2.5, 3.5])

plt.show()
```

# Zadanie 2

```python
photos = loadmat('Lista07Dane/ReducedImagesForTraining.mat').get('images')
```

```python
photos_pca = PCA(n_components=0.8)
photos_pca.fit(photos.T)
components = photos_pca.transform(photos.T)
projected = photos_pca.inverse_transform(components)
```

```python
fig, axes = plt.subplots(10, 10, figsize=(14,14), subplot_kw={'xticks':[], 'yticks':[]})
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
for i in range(10):
    for j in range(5):
        axes[i, j].imshow(photos.T[15 * i + j, :].reshape(60, 82).T, cmap='gray')

for i in range(10):
    for j in range(5):
        axes[i, j + 5].imshow(projected[15 * i + j, :].reshape(60, 82).T, cmap='gray')
```

# Zadanie 3

```python
train_photos = loadmat('Lista07Dane/ReducedImagesForTraining.mat').get('images')
test_photos = loadmat('Lista07Dane/ReducedImagesForTesting.mat').get('images')
```

```python
dists = euclidean_distances(train_photos.T, test_photos.T)
cloest = np.argmin(dists, axis=0)

predicted_person = np.array([(k - 1) // 5 + 1 for k in cloest])
true_person = np.repeat(np.arange(50), 2)
np.mean(predicted_person == true_person)
```
